##REGRESIÓN LINEAL


```python
#Grafica de dispersion salario Vs experiencia
#primera parte del eje x y la segunda del eje y
y = salarios["Salario"]
x=salarios["Años de experiencia"]
#scatter es una grafica de dispersion el primer valor es el eje x y el segundo es el eje y para generar la grafiaca
plt.scatter(x,y,color="red")

#titulo del grafico
plt.title("salario vs años de experiencia")

#titulo  eje x
plt.xlabel("Años de experiencia")

#titulo  eje y
plt.ylabel("salarios")

plt.show()
```
